```python
import nibabel as nib
import numpy as np
import pandas as pd
import os
import glob
```

```python
# 경로만 지정!
base_dir = '/mnt/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_after_swift_20240403'


def calculate_volume(nii_path):
    img = nib.load(nii_path) 
    data = img.get_fdata()
    header = img.header
    voxel_volume = np.prod(header.get_zooms())
    unique_elements, counts_elements = np.unique(data, return_counts=True)
    volumes = counts_elements * voxel_volume
    return dict(zip(unique_elements.astype(int), volumes))
```

```python
import os
import pandas as pd



volume_data = []

for subject_dir in os.listdir(base_dir):
    full_subject_dir = os.path.join(base_dir, subject_dir)
    if not os.path.isdir(full_subject_dir) or 'FS222' not in subject_dir:
        continue
    
    double_subject_dir = os.path.join(full_subject_dir, subject_dir)
    if not os.path.isdir(double_subject_dir):
        continue
    
    mri_dir = os.path.join(double_subject_dir, 'mri')
    nii_path = os.path.join(mri_dir, 'aparc.DKTatlas+aseg.deep.nii.gz')
    
    if os.path.isfile(nii_path):
        try:
            vol_data = calculate_volume(nii_path)
            vol_data["subject"] = subject_dir
            volume_data.append(vol_data)
        except OSError as e:
            print(f"Error processing file {nii_path}, skipped  Error: {e}")
    else:
        print(f"File not found, skipped: {nii_path}")

volume_df = pd.DataFrame(volume_data)
```

```python
def load_freesurfer_lut(lut_path):
    lut = {}
    with open(lut_path, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):
                parts = line.split()
                if parts[0].isdigit():
                    lut[int(parts[0])] = ' '.join(parts[1:])
    return lut

freesurfer_lut = load_freesurfer_lut(lut_path)
```

```python
from scipy import stats
import numpy as np
from scipy.stats import sem, t
from numpy import sqrt
```

```python
def calculate_mean_ci(group):

    cv_data = group['CV']
    n = len(cv_data) 
    mean_cv = cv_data.mean()  
    se = sem(cv_data, nan_policy='omit')  
    h = se * t.ppf((1 + 0.95) / 2., n-1) 

    return pd.Series({
        'Mean_CV': mean_cv,
        'CI_Lower': mean_cv - h,
        'CI_Upper': mean_cv + h
    })
```

```python
roi_mean_ci = grouped.groupby('ROI').apply(calculate_mean_ci).reset_index()

roi_mean_ci
```

```python
import pandas as pd
import pingouin as pg


def calculate_icc_for_roi(df):

    icc_results = []


    for roi in df['ROI'].unique():

        df_subset = df[df['ROI'] == roi]
        

        icc = pg.intraclass_corr(data=df_subset, targets='Subject_ID', raters='Session', ratings='Volume', nan_policy='omit')
        

        icc_row = icc[icc['Type'] == 'ICC3k'].iloc[0]
        icc_results.append({
            'ROI': roi,
            'ICC': icc_row['ICC'],
            '95% CI': icc_row['CI95%'],
            'F': icc_row['F'],
            'df1': icc_row['df1'],
            'df2': icc_row['df2'],
            'p-value': icc_row['pval']
        })

    icc_df = pd.DataFrame(icc_results)

    return icc_df

```

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_rel

# Assuming all_data_long is already loaded
unique_sessions = all_data_long['Session'].unique()

results = []

# Iterate over each ROI
for roi in all_data_long['ROI'].unique():
    roi_data = all_data_long[all_data_long['ROI'] == roi]
    
    # Iterate over each subject within this ROI
    for subject in roi_data['Subject_ID'].unique():
        subject_data = roi_data[roi_data['Subject_ID'] == subject]
        all_volumes = subject_data['Volume'].dropna()  # Collect all volumes across sessions

        if len(all_volumes) > 1:
            overall_std = all_volumes.std()  # Pooled standard deviation across all sessions
            #print(f"Subject: {subject}, ROI: {roi}, Overall Std: {overall_std}")

            effect_sizes = []

            # Compare all possible session pairs
            for i in range(len(unique_sessions)):
                for j in range(i+1, len(unique_sessions)):
                    session1_data = subject_data[subject_data['Session'] == unique_sessions[i]]
                    session2_data = subject_data[subject_data['Session'] == unique_sessions[j]]

                    if not session1_data.empty and not session2_data.empty:
                        mean1 = session1_data['Volume'].mean()
                        mean2 = session2_data['Volume'].mean()

                        if overall_std > 0:
                            effect_size_d = (mean2 - mean1) / overall_std
                            effect_sizes.append(effect_size_d)
                            #print(f"Effect Size D between {unique_sessions[i]} and {unique_sessions[j]}: {effect_size_d}")
                        else:
                            print("Skipping due to zero overall standard deviation.")

            # Calculate and store the mean effect size for this subject and ROI
            if effect_sizes:
                mean_effect_size_d = np.mean(effect_sizes)
                results.append({
                    'Subject_ID': subject,
                    'ROI': roi,
                    'Mean Effect Size D': mean_effect_size_d
                })

# Convert results to DataFrame and print
results_df = pd.DataFrame(results)
```

```python
import pandas as pd
from scipy import stats

def calculate_session_mrd_ci(group):
    group = group.sort_values(by='Session')
    volumes = group['Volume'].values

    n = len(volumes)
    if n == 3 or n == 4:
        mean_volume = volumes.mean()
        num_pairs = n * (n - 1) / 2

        diffs = [abs(volumes[i] - volumes[j]) for i in range(n) for j in range(i+1, n)]
        mrd = sum(diffs) / (num_pairs * mean_volume) * 100

        se_mrd = stats.sem(diffs) / (num_pairs * mean_volume) * 100
        ci_95_mrd = stats.t.interval(0.95, df=len(diffs)-1, loc=mrd, scale=se_mrd)

        return pd.Series({'MRD': mrd, 'MRD_CI_Lower': ci_95_mrd[0], 'MRD_CI_Upper': ci_95_mrd[1]})
    else:
        return pd.Series({'MRD': pd.NA, 'MRD_CI_Lower': pd.NA, 'MRD_CI_Upper': pd.NA})
   

final= pd.DataFrame(all_data_long)

mrd_results = final.groupby(['Subject_ID', 'ROI']).apply(calculate_session_mrd_ci).reset_index()
```
