```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 파일 경로
file_path = '/home/limseoyoung/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_before_swift_20240405/OAS1_0001_MR1_mpr-3_Tag_pr100_sr100_ppf100_spf100_ipf100_sipf179.nii.gz'

# NIfTI 파일 읽기
nii_img = nib.load(file_path)

# 이미지 데이터 추출
image_data = nii_img.get_fdata()
# 각 방향의 중앙 슬라이스 계산
sagittal_slice = image_data.shape[0] // 2
coronal_slice = image_data.shape[1] // 2
axial_slice = image_data.shape[2] // 2

# 중앙 슬라이스 시각화
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Sagittal 슬라이스
axes[0].imshow(image_data[sagittal_slice, :, :], cmap='gray', origin='lower')
axes[0].set_title('Sagittal Plane')
axes[0].axis('off')

# Coronal 슬라이스
axes[1].imshow(image_data[:, coronal_slice, :], cmap='gray', origin='lower')
axes[1].set_title('Coronal Plane')
axes[1].axis('off')

# Axial 슬라이스
axes[2].imshow(image_data[:, :, axial_slice], cmap='gray', origin='lower')
axes[2].set_title('Axial Plane')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import nibabel as nib

def plot_central_slices(image_data, title):
    """Plot the central slices of a 3D image data on Sagittal, Coronal, and Axial planes."""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
   
    c_x, c_y, c_z = image_data.shape[0] // 2, image_data.shape[1] // 2, image_data.shape[2] // 2
    

    axes[0].imshow(image_data[c_x, :, :], cmap='gray')
    axes[0].set_title('Sagittal Plane')
    axes[0].axis('off')
    
    axes[1].imshow(image_data[:, c_y, :], cmap='gray')
    axes[1].set_title('Coronal Plane')
    axes[1].axis('off')
    
    axes[2].imshow(image_data[:, :, c_z], cmap='gray')
    axes[2].set_title('Axial Plane')
    axes[2].axis('off')
    

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

file_path_before ='/home/limseoyoung/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_before_swift_20240405/OAS1_0001_MR1_mpr-3_Tag_pr100_sr100_ppf100_spf100_ipf100_sipf179.nii.gz'# Update to your file path
image_data_before = nib.load(file_path_before).get_fdata()


file_path_after = '/home/limseoyoung/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_after_swift_20240403/OAS1_0001_MR1_mpr-3_Tag_pr100_sr100_ppf100_spf100_ipf100_sipf179.nii.gz'     # Update to your file path
image_data_after = nib.load(file_path_after).get_fdata()


plot_central_slices(image_data_before, "Before Reconstruction")


plot_central_slices(image_data_after, "After Reconstruction")
```
