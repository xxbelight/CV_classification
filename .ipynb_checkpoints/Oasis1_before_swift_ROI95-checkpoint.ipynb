```python
import nibabel as nib
import numpy as np
import pandas as pd
import os
import glob
```

```python

base_dir = '/mnt/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_before_swift_20240405'


def calculate_volume(nii_path):
    img = nib.load(nii_path) 
    data = img.get_fdata()
    header = img.header
    voxel_volume = np.prod(header.get_zooms())
    unique_elements, counts_elements = np.unique(data, return_counts=True)
    volumes = counts_elements * voxel_volume
    return dict(zip(unique_elements.astype(int), volumes))
```

```python
import os
import pandas as pd



volume_data = []

for subject_dir in os.listdir(base_dir):
    full_subject_dir = os.path.join(base_dir, subject_dir)
    if not os.path.isdir(full_subject_dir) or 'FS222' not in subject_dir:
        continue
    
    double_subject_dir = os.path.join(full_subject_dir, subject_dir)
    if not os.path.isdir(double_subject_dir):
        continue
    
    mri_dir = os.path.join(double_subject_dir, 'mri')
    nii_path = os.path.join(mri_dir, 'aparc.DKTatlas+aseg.deep.nii.gz')
    
    if os.path.isfile(nii_path):
        try:
            vol_data = calculate_volume(nii_path)
            vol_data["subject"] = subject_dir
            volume_data.append(vol_data)
        except OSError as e:
            print(f"Error processing file {nii_path}, skipped  Error: {e}")
    else:
        print(f"File not found, skipped: {nii_path}")

volume_df = pd.DataFrame(volume_data)
```

```python
# Path to the FreeSurfer Look-Up Table (LUT)
lut_path = '/mnt/drwelly1/Team_Planning/Brain_Volumetry/Algorithms/freesurfer/FreeSurferColorLUT.txt'
```

```python
def load_freesurfer_lut(lut_path):
    lut = {}
    with open(lut_path, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):
                parts = line.split()
                if parts[0].isdigit():
                    lut[int(parts[0])] = ' '.join(parts[1:])
    return lut

freesurfer_lut = load_freesurfer_lut(lut_path)
```

```python
# 모든 컬럼 이름을 변경
new_columns = [freesurfer_lut.get(int(col), 'Unknown') if col != 'subject' else 'subject' for col in volume_df_mpr1 .columns]


volume_df_mpr1 .columns = new_columns
```

```python

def extract_number(subj):
    parts = subj.split('_')  # '_'를 기준으로 문자열 분리
    if len(parts) > 1:
        number_part = parts[1]  # "OAS1_" 다음 오는 부분을 선택
        # 숫자만 추출
        numbers = ''.join(filter(str.isdigit, number_part))
        return int(numbers) if numbers.isdigit() else None
    return None
```

```python
import pandas as pd
from scipy import stats
import numpy as np


roi_variance_within_subject = all_data_long.groupby(['Subject_ID', 'ROI'])['Volume'].var()

mean_variance_per_subject = roi_variance_within_subject.groupby('Subject_ID').mean()


pivot_table = roi_variance_within_subject.unstack(level='ROI')
cv_values = pivot_table.std() / pivot_table.mean()

n = len(all_data_long['Subject_ID'].unique())
se_cv = cv_values / np.sqrt(n)


confidence_interval = 1.96 * se_cv  # 95% 신뢰 구간을 위한 Z 값(1.96) 사용


cov_df1 = pd.DataFrame({
    'ROI': cv_values.index,
    'CV': cv_values.values,
    'CI_Lower': cv_values.values - confidence_interval,
    'CI_Upper': cv_values.values + confidence_interval
}).reset_index(drop=True)

```

```python
import pandas as pd
from scipy import stats

# Assuming `all_data_long` is your DataFrame with columns: ['Subject', 'ROI', 'Volume', 'Session', 'Subject_ID']
# and that df is already loaded into the environment.

# Define a function to calculate MRD and its confidence interval for each subject's sessions
def calculate_session_mrd_ci(group):
    # Ensure the group is sorted by session to correctly pair the session volumes
    group = group.sort_values(by='Session')
    volumes = group['Volume'].values
    
    # Check if there are four sessions for calculation
    if len(volumes) == 4:
        mean_volume = volumes.mean()
        # Calculate the MRD according to the formula for four sessions
        diffs = [abs(volumes[i] - volumes[j]) for i in range(len(volumes)) for j in range(i+1, len(volumes))]
        mrd = sum(diffs) / (6 * mean_volume) * 100
        
        # Calculate the standard error of MRD
        se_mrd = stats.sem(diffs) / (6 * mean_volume) * 100
        
        # 95% confidence interval for MRD
        ci_95_mrd = stats.t.interval(0.95, df=len(diffs)-1, loc=mrd, scale=se_mrd)
        
        return pd.Series({'MRD': mrd, 'MRD_CI_Lower': ci_95_mrd[0], 'MRD_CI_Upper': ci_95_mrd[1]})
    else:
        # Return NaN or some indicator if there are not exactly four sessions
        return pd.Series({'MRD': pd.NA, 'MRD_CI_Lower': pd.NA, 'MRD_CI_Upper': pd.NA})

# Apply the function to each group of 'Subject_ID' and 'ROI'
mrd_results = all_data_long.groupby(['Subject_ID', 'ROI']).apply(calculate_session_mrd_ci).reset_index()

# Now mrd_results DataFrame will have the MRD and its CIs for each 'Subject_ID' and 'ROI'
# To get the average MRD and CIs per ROI, we can average over all subjects within each ROI
final_results = mrd_results.groupby('ROI').agg({
    'MRD': 'mean',
    'MRD_CI_Lower': 'mean',
    'MRD_CI_Upper': 'mean'
}).reset_index()
```
