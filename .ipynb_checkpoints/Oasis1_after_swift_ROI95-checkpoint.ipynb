```python
import nibabel as nib
import numpy as np
import pandas as pd
import os
import glob
```

```python
# 경로만 지정!
base_dir = '/mnt/drwelly1/Team_Planning/Brain_Volumetry/OASIS1_after_swift_20240403'


def calculate_volume(nii_path):
    img = nib.load(nii_path) 
    data = img.get_fdata()
    header = img.header
    voxel_volume = np.prod(header.get_zooms())
    unique_elements, counts_elements = np.unique(data, return_counts=True)
    volumes = counts_elements * voxel_volume
    return dict(zip(unique_elements.astype(int), volumes))
```

```python
import os
import pandas as pd



volume_data = []

for subject_dir in os.listdir(base_dir):
    full_subject_dir = os.path.join(base_dir, subject_dir)
    if not os.path.isdir(full_subject_dir) or 'FS222' not in subject_dir:
        continue
    
    double_subject_dir = os.path.join(full_subject_dir, subject_dir)
    if not os.path.isdir(double_subject_dir):
        continue
    
    mri_dir = os.path.join(double_subject_dir, 'mri')
    nii_path = os.path.join(mri_dir, 'aparc.DKTatlas+aseg.deep.nii.gz')
    
    if os.path.isfile(nii_path):
        try:
            vol_data = calculate_volume(nii_path)
            vol_data["subject"] = subject_dir
            volume_data.append(vol_data)
        except OSError as e:
            print(f"Error processing file {nii_path}, skipped  Error: {e}")
    else:
        print(f"File not found, skipped: {nii_path}")

volume_df = pd.DataFrame(volume_data)
```

```python
lut_path = '/mnt/drwelly1/Team_Planning/Brain_Volumetry/Algorithms/freesurfer/FreeSurferColorLUT.txt'
# labeling 불러오면됨
```

```python
def load_freesurfer_lut(lut_path):
    lut = {}
    with open(lut_path, 'r') as file:
        for line in file:
            if line.strip() and not line.startswith('#'):
                parts = line.split()
                if parts[0].isdigit():
                    lut[int(parts[0])] = ' '.join(parts[1:])
    return lut

freesurfer_lut = load_freesurfer_lut(lut_path)
```

```python
# 모든 컬럼 이름을 변경
new_columns = [freesurfer_lut.get(int(col), 'Unknown') if col != 'subject' else 'subject' for col in volume_df_mpr1 .columns]


volume_df_mpr1 .columns = new_columns
```

```python

def extract_number(subj):
    parts = subj.split('_')  # '_'를 기준
    if len(parts) > 1:
        number_part = parts[1]  
        # 숫자만 
        numbers = ''.join(filter(str.isdigit, number_part))
        return int(numbers) if numbers.isdigit() else None
    return None
```

# left and right 맞춰서 다시 정리해서 cov 값 뽑아내기!

```python

brain_areas = set()
left_right_columns = []

for col in volume_df_sorted1.columns:
    if col.startswith('Left-') or col.startswith('Right-'):
        left_right_columns.append(col)
        brain_area = col.split('-', 1)[1]
        brain_areas.add(brain_area)


for area in brain_areas:
    left_col = f'Left-{area}'
    right_col = f'Right-{area}'
    volume_df_sorted1[area] = volume_df_sorted1.get(left_col, 0) + volume_df_sorted1.get(right_col, 0)

volume_df_sorted1_final = volume_df_sorted1.drop(columns=[col for col in left_right_columns if col not in ['subject']])
```

```python
brain_areas_ctx = set()
ctx_columns = []

for col in volume_df_sorted1_final.columns:
    if col.startswith('ctx-lh-') or col.startswith('ctx-rh-'):
        ctx_columns.append(col)
     
        brain_area_ctx = col.split('-', 2)[-1]
        brain_areas_ctx.add(brain_area_ctx)

for area in brain_areas_ctx:
    lh_col = f'ctx-lh-{area}'
    rh_col = f'ctx-rh-{area}'
   
    volume_df_sorted1_final[area] = volume_df_sorted1_final[lh_col].fillna(0) + volume_df_sorted1_final[rh_col].fillna(0)


volume_df_sorted1_final = volume_df_sorted1_final.drop(columns=ctx_columns)
```

# cv구하기 전 전처리 

```python

all_data = pd.concat([volume_df_sorted1_final ,volume_df_sorted2_final ,volume_df_sorted3_final ,volume_df_sorted4_final])
all_data_long = pd.melt(all_data, id_vars=["subject"], var_name="ROI", value_name="Volume")
```

```python

all_data_long['Session'] = all_data_long['subject'].str.extract(r'(mpr-\d)')

all_data_long['ROI'] = all_data_long['ROI'].str.extract(r'([a-zA-Z-_]+)')[0]
```

```python

all_data_long['Subject_ID'] = all_data_long['subject'].str.extract(r'(OAS1_\d+)')
all_data_long['Session'] = all_data_long['subject'].str.extract(r'(mpr-\d)')

```

```python
roi_variance_within_subject = all_data_long.groupby(['Subject_ID', 'ROI'])['Volume'].var()


mean_variance_per_subject = roi_variance_within_subject.groupby('Subject_ID').mean()

```

```python
pivot_table = roi_variance_within_subject.unstack(level='ROI')
```

```python
cov_values = pivot_table.std() / pivot_table.mean()


cov_df = pd.DataFrame(cov_values, columns=['Coefficient_of_Variation']).reset_index()


cov_df.rename(columns={'index': 'ROI'}, inplace=True)
```
